```python
%pylab 
import pandas as pd
import glob
import re
import os
from collections import defaultdict
import sys
if sys.version_info[0] < 3:
    from StringIO import StringIO
else:
    from io import StringIO
```

```python
frex =re.compile(r"NodeR-parsed-cesson-(\w+)-(SF\d+)\-(2016\-03\-06\ \d+\:\d+\:\d+\-2016\-03\-06\ \d+\:\d+\:\d+).csv")
files = ['/home/john/parsingSunday/'+f for f in os.listdir('/home/john/parsingSunday/') if frex.match(f)]
time_place = {"10:40":"cessonGolf", "11:55":"La Terasse", "12:45":"Joualt Hubert farm", "14:02":"Maison Medicale","16:00":"Parc Hamelin","16:35":"Cemeterie Est",
              "17:15":"Frebardiere","18:05":"Parc du Bois"}
print files
nname = "%s, PER=%s%%"
lsort = []
#inner_dict = defaultdict(lambda x: defaultdict(list))
sf_dict = defaultdict(lambda : defaultdict(list))
pl_dict = defaultdict(lambda: defaultdict(list))
per_dict = defaultdict(dict)
time_dict = defaultdict(dict)
stat_dict = defaultdict(dict)
lsort = [frex.search(f).groups()[2] for f in files]
for f in files:
    matched = frex.search(f).groups()
    sf = matched[1]
    place = matched[2]
    for k,v in time_place.iteritems():
        if k in place:
            place = v
    print place
    lsort.append(place)
    rd = pd.read_csv(f)
    per=float(len(rd.index))/500.0
    per = 1-per
    per = per*100
    sf_dict[sf][nname % (place,per)]=(rd.rssi)
    pl_dict[place][nname % (sf,per)] =rd.rssi
    time_dict[sf][nname % (place,per)] =pd.to_datetime(rd.time)
    #sf_dict[sf][nname % (place,per)]=(rd.lsnr)
    #pl_dict[place][nname % (sf,per)] =rd.lsnr
    sf_dict[sf][nname % (place,per)]=(rd.lsnr/rd.rssi)
    pl_dict[place][nname % (sf,per)] =rd.lsnr/rd.rssi
    per_dict[sf][place] = per
    stat_dict[place][sf] = [rd.rssi.mean(),rd.rssi.std(),rd.lsnr.mean(),rd.lsnr.std(),"%s%%" %per,"%s/%s" %(len(rd.index),500)]
    
print per_dict
print stat_dict
joined ="Point,SF,SNR(dB) mean, SNR std. deviation, RSSI(dBm) mean, RSSI std. deviation,PER, Packets Rx/Tx\n"
for  pl,d in stat_dict.iteritems():
    for sf,stats in d.iteritems():
        st= ",".join(str(it) for it in stats)
        joined+= "%s,%s,%s\n" %(pl,sf,st)
        print joined
        print sf + pl
print joined

pd.read_csv(StringIO(joined)).to_csv("/home/john/parsingSunday/table.csv")
    

```
